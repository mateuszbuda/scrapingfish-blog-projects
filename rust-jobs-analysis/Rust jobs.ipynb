# Scraping Rust job offers

## Indeed

```python
job_offers = []
```

```python
def get_description(job_url):
    url = quote_plus(job_url)
    response = requests.get(f"{url_prefix}{url}")
    soup = BeautifulSoup(response.content, "html.parser")
    return soup.select("div#jobDescriptionText")[0].text
```

```python
job_selector = "ul.jobsearch-ResultsList > li > div.cardOutline"

for offset in range(210, 400, 10):
    url = quote_plus(f"https://www.indeed.com/jobs?q=rust&start={offset}")
    response = requests.get(f"{url_prefix}{url}")
    soup = BeautifulSoup(response.content, "html.parser")

    jobs_elements = soup.select(job_selector)
    for job_element in jobs_elements:
        offer_id = job_element.select("h2.jobTitle a")[0]["data-jk"]
        job_url = f"https://www.indeed.com/viewjob?jk={offer_id}"
        job_title = job_element.select("table.jobCard_mainContent h2.jobTitle")[0].text
        job_location = job_element.select("div.companyLocation")[0].text
        job_location = re.sub(r"\+\d+\slocation(s){0,1}", "", job_location)
        remote = "no"
        if "hybrid remote" in job_location.lower():
            remote = "hybrid"
        elif "remote" in job_location.lower():
            remote = "yes"
        job_location = job_location.replace("Remote in ", "").replace("Hybrid remote", "")
        job_description = get_description(job_url)

        job = {
            "title": job_title,
            "url": job_url,
            "location": job_location,
            "remote": remote,
            "description": job_description
        }
        print(job_title, job_url, job_location, remote)
        job_offers.append(job)
```

# justjoin.it

```python
url = "https://justjoin.it/?q=Rust@skill"
url = quote_plus(url)
url_prefix = f"https://scraping.narf.ai/api/v1/?render_js=false&api_key={API_KEY}&url="
response = requests.get(f"{url_prefix}{url}")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
urls = [f"https://justjoin.it{el['href']}" for el in soup.select('a[href*="offers"]')]
titles = [el.text for el in soup.select('a[href*="offers"]+div>div:last-child>div:first-child>div:first-child>div')]
locations_and_remote = [el.text for el in soup.select('a[href*="offers"]+div>div:last-child>div:last-child>div:nth-child(2)>div:nth-child(2)>span') if not el.text.startswith(", ")]
locations = []
remote = []
for i in range(0, len(locations_and_remote), 2):
    locations.append(locations_and_remote[i].split(',')[0])
    remote_type = locations_and_remote[i+1]
    if remote_type.lower() == "fully remote":
        remote.append("yes")
    else:
        remote.append("no")
        
salary_ranges = [el.text for el in soup.select('a[href*="offers"]+div>div:last-child>div:first-child>div:nth-child(2)')]
salary_mins = []
salary_maxs = []
for salary_range in salary_ranges:
    if salary_range.lower() == "undisclosed salary":
        salary_mins.append(None)
        salary_maxs.append(None)
        continue
        
    salary_min, salary_max = salary_range.replace("PLN", "").split(" - ")
    salary_mins.append(int(salary_min.replace(" ", "").strip()))
    salary_maxs.append(int(salary_max.replace(" ", "").strip()))
```

```python
for job_url, job_title, job_location, job_remote, job_salary_min, job_salary_max in zip(urls, titles, locations, remote, salary_mins, salary_maxs):
    description_url = quote_plus(job_url)
    response = requests.get(f"{url_prefix}{description_url}")
    description_soup = BeautifulSoup(response.content, "html.parser")
    job_description = description_soup.select("div#root div.MuiPaper-root")[1].text
    job_offers.append({
            "title": job_title,
            "url": job_url,
            "location": job_location,
            "remote": job_remote,
            "description": job_description,
            "salary_min": job_salary_min,
            "salary_max": job_salary_max,
            "salary_currency": "PLN",
        })
```

## rustjobs.dev
A lot of postings here are quite old (50+ days) but it's still relevant for our estimates.
*Requires JavaScript rendering*

```python
url = quote_plus("https://rustjobs.dev/")
url_prefix = f"https://scraping.narf.ai/api/v1/?render_js=true&api_key={API_KEY}&url="
response = requests.get(f"{url_prefix}{url}")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
all_lists = soup.select("div > ul")
featured = all_lists[0]
rest = all_lists[2]
urls = [anchor["href"] for anchor in featured.select("h2 > a")] + [anchor["href"] for anchor in rest.select("div > a") if anchor.text.lower() != 'remote']
urls = [f"https://rustjobs.dev/featured-jobs{url}" for url in urls]
titles = [anchor.text for anchor in featured.select("h2 > a")] + [anchor.text for anchor in rest.select("div > a") if anchor.text.lower() != 'remote']
countries = [span.text for span in featured.select("div > div > div > div > div:last-child > svg+span")] + [span.text for span in rest.select("div > div > div > div > div:last-child > svg+span")]
cities = [span.text for span in featured.select("div > div > div > div > div:nth-last-child(2) > svg+span")] + [span.text for span in rest.select("div > div > div > div > div:nth-last-child(2) > svg+span")]
# TODO - store cities AND countries?
locations = cities
remote = ["yes"] * len(cities) # all jobs there are remote

salary_mins = []
salary_maxs = []
for li in featured.select("li"):
    salary_element = li.select("div > div > div > div > div > div:nth-last-child(3)")
    if salary_element and "$" in salary_element[0].text:
        salary_min, salary_max = salary_element[0].text.split(" - ")
        salary_mins.append(1000 * int(salary_min.replace("$", "").replace("K", "")) / 12)
        salary_maxs.append(1000 * int(salary_max.replace("$", "").replace("K", "")) / 12)
    else:
        salary_mins.append(None)
        salary_maxs.append(None)
```

```python
job_offers = []
for job_url, job_title, job_location, job_remote, job_salary_min, job_salary_max in zip(urls, titles, locations, remote, salary_mins, salary_maxs):
    description_url = quote_plus(job_url)
    response = requests.get(f"{url_prefix}{description_url}")
    description_soup = BeautifulSoup(response.content, "html.parser")
    job_description = description_soup.select("div.markdown-component.prose")[0].text
    job_offers.append({
            "title": job_title,
            "url": job_url,
            "location": job_location,
            "remote": job_remote,
            "description": job_description,
            "salary_min": job_salary_min,
            "salary_max": job_salary_max,
            "salary_currency": "USD"
        })
```

# Startup jobs

```python
url_prefix = f"https://scraping.narf.ai/api/v1/?api_key={API_KEY}&url="
```

```python
# results after 6th page don't have actual rust job offers
for page in range(1, 7):
    page_url = quote_plus(f"https://startup.jobs/?q=Rust&page={page}")
    response = requests.get(f"{url_prefix}{page_url}")
    soup = BeautifulSoup(response.content, "html.parser")
    hits = soup.select("div[data-search-target='hits'] > div.postCard")
    for offer in hits:
        title_el = offer.select("a.postCard__title")[0]
        url = f'https://startup.jobs{title_el["href"]}'
        title = title_el.text
        location = offer.select(".postCard__location")[0].text
        remote = "yes" if location.lower().strip() == "remote" else "no"

        response = requests.get(f"{url_prefix}{url}")
        description_soup = BeautifulSoup(response.content, "html.parser")
        description = description_soup.select(".jobListing__main__text")[0].text

        job_offers.append({
            "url": url,
            "title": title.strip(),
            "location": location.strip(),
            "remote": remote,
            "salary_min": None,
            "salary_max": None,
            "salary_currency": "USD",
            "description": description.strip()
        })
```
